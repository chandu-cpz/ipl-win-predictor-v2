```python

def get_umpire3_fill_value(df):
    """
    Returns the most frequent non-null value in the 'umpire3' column
    that is different from 'umpire1' and 'umpire2'.
    """
    # Get the value counts of non-null values in 'umpire3'
    umpire3_counts = df['umpire3'].value_counts(dropna=True)

    # Iterate over the value counts to find the first value that is different from 'umpire1' and 'umpire2'
    for umpire, count in umpire3_counts.items():
        if (df['umpire1'] != umpire).any() and (df['umpire2'] != umpire).any():
            return umpire

    # If no suitable value is found, return None
    return None

fill_value = get_umpire3_fill_value(df)

if fill_value is not None:
    df['umpire3'] = df['umpire3'].fillna(fill_value)
else:
    print("No suitable value found for filling 'umpire3'")
```

# Custom IQR Removal Transformer

```python
class Outlier_Remover(BaseEstimator, TransformerMixin):
    
    def __init__(self, action='keep'): 
        self.action = action
        
    def fit(self, X, y=None):
        self.median_ = np.median(X, axis=0)
        return self
    
    def transform(self, X):
        Q1 = np.percentile(X, 25, axis=0)
        Q3 = np.percentile(X, 75, axis=0) 
        
        IQR = Q3 - Q1

        lower = Q1 - 1.5*IQR
        upper = Q3 + 1.5*IQR
        
        outlier_mask = (X < lower) | (X > upper)
        
        if self.action == 'drop':
            return X[~outlier_mask]
        else:
            for i in range(X.shape[1]):
                X[:, i][outlier_mask[:, i]] = self.median_[i]
            return X
```
